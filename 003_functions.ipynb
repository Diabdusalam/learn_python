# Fungsi

Fungsi adalah blok kode yang dapat dipanggil untuk melakukan tugas tertentu. Ia dapat menerima input dan menghasilkan output. Ini adalah cara untuk mengatur kode Anda agar dapat digunakan kembali. Sintaks untuk mendefinisikan fungsi adalah sebagai berikut:

```python
def add(a, b):
    return a + b

print(add(2, 3))
add(9, 14)
```

Python adalah bahasa yang sensitif terhadap indentasi. Jika kamu mau masukkan blok kode masuk ke dalam sebuah fungsi, kamu perlu mengidentasi blok kode tersebut dengan tab.

Jadi, selama:
- Kode setelah definisi fungsi diindentasi, itu dianggap sebagai bagian dari fungsi.
- Setiap kode setelah itu yang berada pada tingkat indentasi yang sama, selama tidak dipisahkan oleh baris kosong, dianggap sebagai bagian dari fungsi yang sama

Jadi kita dapat memiliki beberapa baris kode seperti ini:

```python
def several_text_manipulation(text):
    data = text.replace('cat', 'dog')
    data = data.upper()
    return data

several_text_manipulation("Hi, I'm a cat")
```

Jadi seperti yang bisa kita lihat di atas, kita memiliki satu parameter `text` (kita akan membahas istilah ini secara detail sebentar lagi), lalu kita memanipulasinya di dalam fungsi, setelah itu kita mengembalikan hasilnya, yang akan menjadikannya output dari fungsi.

```python
def several_text_manipulation(text):
  data = text.replace('cat', 'dog')
  data = data.upper()
  return data

several_text_manipulation("Hi, I'm a cat")
```

# Intermezo: Parameter? Argumen?

> Bagian ini tidak wajib untuk dipahami, ini untuk beberapa dari Anda yang mungkin belajar di luar kursus ini dan mungkin pernah mendengar istilah ini sebelumnya. Selama sisa kursus ini Anda dapat berpura-pura bahwa kedua istilah ini dapat dipertukarkan, dan kursus ini akan menggunakan istilah "parameter" untuk keduanya.

Kita akan terus mendengar kedua istilah ini ketika berbicara tentang fungsi. Jadi apa itu dan apa bedanya?

## Saat kita menulis fungsi itu disebut parameter

Saat kita menulis fungsi, kita dapat mendefinisikan parameter fungsi tersebut. Jumlah total parameter dapat berbeda dari satu fungsi ke fungsi lain sesuai dengan kebutuhan fungsi, terkadang bisa nol, terkadang bisa satu, terkadang bisa lebih dari satu.

```python
def sum_three(a, b, c):
    return a + b + c
```

Dari contoh di atas, yang disebut sebagai parameter adalah `a`, `b`, dan `c`.

## Saat kita menggunakan fungsi, kita memberikan argumen

Sekarang ketika kita **menggunakan** fungsi setelah fungsi tersebut didefinisikan, kita memberikan argumen ke fungsi tersebut. Jadi jika kita menggunakan fungsi `sum_three` di atas, kita dapat menggunakannya dengan `3`, `4`, dan `5` sebagai argumen:

```python
sum_three(3, 4, 5)
```

Tentu saja kita dapat menggunakan variabel sebagai argumen:

```python
num_1 = 3
num_3 = 5

sum_three(num_1, 4, num_3)
```

Sulit untuk mengingat antara parameter dan argumen? Untuk pembelajaran, Anda dapat mengingat bahwa kedua kata itu sangat mirip, dan tidak mengetahui mana yang mana bukanlah masalah besar, tetapi Anda akan menemukan kedua kata itu banyak digunakan di sini dan sumber pembelajaran lainnya, tetapi Anda dapat berpura-pura bahwa mereka adalah hal yang sama.

# Argumen Posisi vs Kata Kunci

Perhatikan fungsi di bawah ini:

```python
def substract(a, b):
    return a - b
```

Seperti yang kami katakan sebelumnya, kita dapat menggunakannya seperti di bawah ini:

```python
substract(5, 2)
```

Ini disebut **argumen posisi**. Argumen diteruskan ke fungsi sesuai dengan posisi parameter dalam definisi fungsi.

Cara lain untuk meneruskan argumen adalah dengan menggunakan nama parameter:

```python
substract(a=5, b=2)
```

Ini disebut **argumen kata kunci**. Argumen diteruskan ke fungsi menggunakan nama parameter. Yang membedakan argumen kata kunci dari argumen posisi adalah bahwa **urutan argumen tidak masalah**, yang berarti kita dapat melakukan ini dan hasilnya akan sama:

```python
substract(b=2, a=5)
```
